```python
from RegressionTree import RegressionTree
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from GBDT import GradientBoostingRegressor, GradientBoostingClassifier
```

```python
boston = load_boston()
X, y = boston.data, boston.target
feature_name = boston.feature_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train_df = pd.DataFrame(X_train, columns=feature_name)
y_train_df = pd.Series(y_train)
print(X_train_df.head(5))
print(y_train_df.head(5))
```

```python
model = GradientBoostingRegressor(3,4)
model.fit(X_train_df, y_train_df)
```

```python
X_test_df = pd.DataFrame(X_test, columns=feature_name)
y_test_df = pd.Series(y_test)
```

```python
y_pred = model.predict(X_test_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))
plt.plot(X_test_df.index, y_test_df.values)
plt.plot(X_test_df.index, y_pred)
```
